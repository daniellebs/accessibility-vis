```python
DATA_PATH = '../../input_data/GTFS-28-Oct-19/'
OUTPUT_PATH = '../../output_data/'
```

```python
FILES_PREFIX = 'morning'
file_name = OUTPUT_PATH + FILES_PREFIX + '_nodes.pkl'
print(f'Loading nodes from file {file_name}')
nodes_df = pd.read_pickle(file_name)
nodes_df.head(3)
```

```python
nodes_df[nodes_df['node_id'] == 133828]
```

```python
example_lengths_df = sp_5_df[sp_5_df['source'] == 133828]
```

```python
example_lengths_df.head(3)
```

```python
example_lengths_df.shape
```

```python
nodes_small_df = nodes_df[['node_id', 'stop_lon', 'stop_lat', 'arrival', 'departure', 'trip_id', 'stop_id']]
```

```python
dists_from_example_source_df = nodes_small_df.merge(example_lengths_df, left_on='node_id', right_on='target')
```

```python
dists_from_example_source_df
```

```python
dists_from_example_source_df = nodes_small_df[['node_id', 'departure', 'stop_lon', 'stop_lat', 'trip_id', 'stop_id']].merge(dists_from_example_source_df, left_on='node_id', right_on='source', suffixes=('_source', ''))
```

```python
dists_from_example_source_df.head(3)
```

```python
dists_from_example_source_df = dists_from_example_source_df.drop(['node_id_source', 'node_id', 'departure'], axis=1)
```

```python
dists_from_example_source_df.head(3)
```

```python
dists_from_example_source_df.to_csv(OUTPUT_PATH + FILES_PREFIX + '_example_node_times.csv')
```

```python
dists_from_example_source_df['arrivals_difference'] = dists_from_example_source_df['arrival'] - dists_from_example_source_df['departure_source']
```

```python
import datetime as dt
dists_from_example_source_df['time'] = dists_from_example_source_df['time_sec'].apply(lambda x: dt.timedelta(seconds=x))
```
